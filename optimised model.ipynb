```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
import pickle
from kerastuner import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameters
import numpy as np
import pickle
```

```python
pickle_in=open("X_train.pickle", "rb")
X_train=pickle.load(pickle_in)
pickle_in.close()

pickle_in=open("X_test.pickle", "rb")
X_test=pickle.load(pickle_in)
pickle_in.close()

pickle_in=open("y_train.pickle", "rb")
Y_train=pickle.load(pickle_in)
pickle_in.close()

pickle_in=open("y_test.pickle", "rb")
Y_test=pickle.load(pickle_in)
pickle_in.close()
```

```python
IMG_CHANNELS=3
IMG_ROWS=32
IMG_COLS=32

BATCH_SIZE=128
NB_EPOCH=20
NB_CLASSES=10
VERBOSE=1
VALIDATION_SPLIT=0.2
```

```python
def build_model(hp):
    model=Sequential()
    model.add(Conv2D(
        filters=hp.Int("conv_1_layer",64,128,32),
        kernel_size=hp.Choice("conv_1_kernel",values=[3,5]),
        activation="relu",
        input_shape=(32,32,3)
    ))
    model.add(Dropout(hp.Choice("Dropout_1", values=[0.2,0.3,0.4])))
    model.add(Conv2D(
        filters=hp.Int("conv_2_layer",32,64,32),
        kernel_size=hp.Choice("conv_2_kernel",values=[3,5]),
        activation="relu"
    ))
    model.add(Dropout(hp.Choice("Dropout_2", values=[0.2,0.3,0.4])))
    model.add(Flatten())
    model.add(Dense(
        units=hp.Int("dense_1_layer",32,128,16),
        activation="relu"
    ))
    model.add(Dense(10,activation="softmax"))
    
    model.compile(
        optimizer=keras.optimizers.Adam(hp.Choice("learning_rate", values=[1e-2,1e-3])),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
    
```

```python
tuner=RandomSearch(
    build_model,
    objective="val_accuracy",
    max_trials=4,
    directory="output3",
    project_name="cifar10"
)
```

```python
tuner.search(X_train, Y_train, epochs=3, validation_split=0.1)
```

```python
model=tuner.get_best_models(num_models=1)[0]
model.summary()
```

```python
model.fit(X_train, Y_train, epochs=20, validation_split=0.1, initial_epoch=3)
```

```python
model.save("revised.h5")
```

```python
score=model.evaluate(X_test, Y_test)
```
